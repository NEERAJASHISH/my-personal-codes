```python
top_ten_tags = []

quotes = requests.get("https://quotes.toscrape.com/page/1/")
soup  = bs4.BeautifulSoup(quotes.text,"lxml")

quotediv = soup.select(".quote")

for k in range(10):
    extract = quotediv[k]
    get = extract.select("a")
    for n in get:
        if "love" in str(n):
            top_ten_tags.append("love")
        elif "inspirational"  in str(n):
            top_ten_tags.append("inspirational")
        elif "life" in str(n):
            top_ten_tags.append("life")
        elif "humor" in str(n):
            top_ten_tags.append("humor")
        elif "books" in str(n):
            top_ten_tags.append("books")
        elif "reading" in str(n):
            top_ten_tags.append("reading")
        elif "friendship" in str(n):
            top_ten_tags.append("friendship")
        elif "friends" in str(n):
            top_ten_tags.append("friends")
        elif "truth" in str(n):
            top_ten_tags.append("truth")
        elif "simile" in str(n):
            top_ten_tags.append("simile")
            
            
print(top_ten_tags)            
            
       
            
            
            
            
            
    
    
        
    
    
    
    
    











```

```python
# this is intelligent code
top_tens_tags = []

quotes = requests.get("https://quotes.toscrape.com/page/1/")
soup  = bs4.BeautifulSoup(quotes.text,"lxml")

quotediv = soup.select(".quote")
tager = soup.select(".tag-item")

for k in range(10):
    extract = quotediv[k]
    get = extract.select("a")
    
    for n in get:
        if str(n) is str(tager):
            top_tens_tags.append(str(tager))
    
    
        
            
print(top_ten_tags)          
```
