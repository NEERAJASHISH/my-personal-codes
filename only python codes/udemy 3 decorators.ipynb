```python
def decorators(func_needs_decoration):
    
    def wrap():
        print("codes before the function!!!!")
        
        func_needs_decoration()
        
        print("codes after the decoration!!!!")
        
    return wrap  
    
   
    
    
    
```

```python
def merge_the_tools(string, k):
    for i in range(0,len(string), k):
    	#slice string upto k characters
        line = string[i:i+k]
        seen = set()
        for i in line:
        	#only print if we haven't already seen this character
            if i not in seen:
                print(i,end="")
                seen.add(i)
        #prints a new line
        print()

if __name__ == '__main__':
    string, k = input(), int(input())
    merge_the_tools(string, k)
```

```python
def strnum(m):
    return [str(i) for i in range(m)]


strnum(10)
    
```

```python
def str2m(m):
    return list(map(str,range(m)))

str2m(10)
```

```python
#current times
start_time = time.time()

#run code
code_run = strnum(1000000)

#current time after running the code
end_time = time.time()

# time after running
total_time  = end_time - start_time

print(total_time)
```
